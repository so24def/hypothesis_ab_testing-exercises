$\mu = 800$, $\sigma^2 = 1600$, $\bar{x} = 820$, n = 36, $\alpha = 0.05$

$ H_0 : \mu = 800 $ olduğu biliniyor.

$H_1$ hipotezi büyüklük içerdiğinden, __tek yanlı hipotez testi__ yapılır ve kritik nokta __pozitif__ taraftadır.


```python
mu = 800
var = 1600
xbar = 820
n = 36
significance = 0.05

z = z_test(mu=mu, var=var, xbar=xbar, sample_size=n, significance=significance,one_tail=True )
```

_Bir sınıftaki kız ve erkek öğrencilerin istatistik genel sınavından aldığı puanlar aşağıdaki gibidir_:

_%99 güven düzeyinde istatistik dersinde kız öğrencilerin erkek öğrencilerden daha başarılı olduğu iddiasını test ediniz._

Çözüm:

Örneklemlerin ortalamaları arasındaki fark için, örneklem hacmi > 30 ise kitlenin normal dağılıma yakınsadığı varsayılıp, örneklemin nokta tahminleri ile z istatistiği kullanılabilir.

* Ayrıca her örneklemin ortalaması ve varyansı hesaplanmalıdır, örneklem istatistikleri kullanılacaktır.

$ H_0 : \mu_k = \mu_e $ kız ve erkeklerin başarılarının eşit olması.

$H_1 : \mu_k > \mu_e $ kızların erkeklerden daha başarılı olduğu iddiası.


$H_1$ hipotezi büyüklük içerdiğinden, __tek yanlı hipotez testi__ yapılır ve kritik nokta __pozitif__ taraftadır.

```python
kız = [81, 67, 90, 78, 81, 80, 76]
erkek = [72, 69, 98, 66, 85, 76, 79, 80, 77]

xbar_kız = np.mean(kız)
var_kız = np.var(kız,ddof=1)

xbar_erkek = np.mean(erkek)
var_erkek = np.var(erkek,ddof=1)

significance= 0.01

t = t_test_diff(xbar_1= xbar_kız, xbar_2= xbar_erkek, 
                var_1= var_kız, var_2= var_erkek, 
                sample_size_1= len(kız), sample_size_2= len(erkek),
                significance=significance, one_tail=True)
```

```python
t_dist_plot(t_value= t, dof= len(kız) + len(erkek) - 2, significance= significance,two_tail=False, positive_tail=True)
```

* Aynı zamanda p değeri, önem değerinden büyük olduğu için, test istatistiğinin kabul bölgesine düştüğünü ve benzer şekilde $H_0$ hipotezinin reddedilmesi için __yeterli kanıt olmadığını gösterir.__

* Aynı zamanda p değeri, önem değerinden küçük olduğu için, test istatistiğinin red bölgesine düştüğünü ve benzer şekilde $H_0$ __hipotezinin reddedilmesi için yeterli kanıt olduğunu gösterir.__

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def z_dist_plot(z_value, significance, two_tail=False, positive_tail=True):
    
    if two_tail == False:
        
        alpha = round(stats.norm.ppf(1-significance),4)
    
        x = np.linspace(-4, 4, 10000)

        plt.plot(x, 1/np.sqrt(2*np.pi) * np.exp(-x**2/2),color='white')

        if positive_tail:

            plt.fill_between(x[x<=alpha], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[x<=alpha]**2/2), color='green')
            plt.fill_between(x[x>=alpha], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[x>=alpha]**2/2), color='red')
            plt.axvline(x=z_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
            plt.text(z_value+0.01, -0.015, f"z-value = {z_value:.3f}", ha='left', fontsize=8)
            
        else:

            plt.fill_between(x[x<=-alpha], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[x<=-alpha]**2/2), color='red')
            plt.fill_between(x[x>=-alpha], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[x>=-alpha]**2/2), color='green')
            plt.axvline(x=z_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
            plt.text(z_value+0.01, -0.015, f"z-value = {z_value:.3f}", ha='left', fontsize=8,color='black')

    else:
        
        alpha = round(stats.norm.ppf(1-significance/2),4)
    
        x = np.linspace(-4, 4, 10000)

        plt.plot(x, 1/np.sqrt(2*np.pi) * np.exp(-x**2/2),color='white')

        plt.fill_between(x[(x<=-alpha)], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[(x<=-alpha)]**2/2), color='red')
        plt.fill_between(x[(x>=alpha)], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[(x>=alpha)]**2/2), color='red')
        plt.fill_between(x[(x>=-alpha) & (x<=alpha)], 0, 1/np.sqrt(2*np.pi) * np.exp(-x[(x>=-alpha) & (x<=alpha)]**2/2), color='green')
        plt.axvline(x=z_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
        plt.text(z_value+0.01, -0.015, f"z-value = {z_value:.3f}", ha='left', fontsize=8)   

    plt.xlabel('Z Values')
    plt.ylabel('Probability Density')
    plt.title(f'Standard Normal Distribution with z-value = {z_value:.3f}')
    plt.show()
    

    
    
def t_dist_plot(t_value, dof, significance, two_tail=False, positive_tail=True):
    
    if two_tail == False:
        
        alpha = round(stats.t.ppf(1-significance, dof),4)
    
        x = np.linspace(-4, 4, 10000)

        plt.plot(x, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x ** 2) / dof)) ** (-(dof+1)/2)),color='white')

        if positive_tail:

            plt.fill_between(x[x<=alpha], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x<=alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='green')
            plt.fill_between(x[x>=alpha], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x>=alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='red')
            plt.axvline(x=t_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
            plt.text(t_value+0.05, -0.005, f"t-value = {t_value:.3f}", ha='left', fontsize=8)
            
        else:

            plt.fill_between(x[x<=-alpha], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x<=-alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='red')
            plt.fill_between(x[x>=-alpha], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x>=-alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='green')
            plt.axvline(x=t_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
            plt.text(t_value+0.05, -0.005, f"t-value = {t_value:.3f}", ha='left', fontsize=8)

    else:
        
        alpha = round(stats.t.ppf(1-significance/2, dof),4)
    
        x = np.linspace(-4, 4, 10000)

        plt.plot(x, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x ** 2) / dof)) ** (-(dof+1)/2)),color='white')

        plt.fill_between(x[(x<=-alpha)], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x<=-alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='red')
        plt.fill_between(x[(x>=alpha)], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[x>=alpha] ** 2) / dof)) ** (-(dof+1)/2)), color='red')
        plt.fill_between(x[(x>=-alpha) & (x<=alpha)], 0, (1 / (np.sqrt(dof) * np.sqrt(np.pi))) * ((1 + ((x[(x>=-alpha) & (x<=alpha)] ** 2) / dof)) ** (-(dof+1)/2)), color='green')
        plt.axvline(x=t_value, ymin=0.035 ,ymax=0.06, color='black', linestyle='-')
        plt.text(t_value+0.05, -0.005, f"t-value = {t_value:.3f}", ha='left', fontsize=8)   

    plt.xlabel('T Values')
    plt.ylabel('Probability Density')
    plt.title(f'T Distribution with t-value = {t_value:.3f}')
    plt.show()
    
    
def t_test(mu, var, xbar, sample_size, significance, one_tail=True):
    
    t_value = (xbar - mu) / np.sqrt(var/sample_size)

    dof = sample_size - 1
        
    if one_tail & (t_value >= 0):
        
        p_value = 1 - stats.t.cdf(t_value, dof)
        
        t_alpha = stats.t.ppf(significance,dof)*(-1)
        
    elif one_tail & (t_value < 0):

        p_value = stats.t.cdf(t_value, dof) 
        
        t_alpha = stats.t.ppf(significance,dof)
        
    if one_tail == False:
        
        p_value = 2*(1 - stats.t.cdf(np.abs(t_value),dof))
        
        t_alpha = stats.t.ppf(significance/2,dof)*(-1)
    
    if p_value > significance:
        print(f't value = {t_value:.5f}')
        print(f't alpha = {t_alpha:.5f}')
        print(f'p value = {p_value:.5f} > {significance} and H0 can not be rejected.')
    else:
        print(f't value = {t_value:.5f}')
        print(f't alpha = {t_alpha:.5f}')
        print(f'p value = {p_value:.5f} <= {significance} and H0 is rejected, H1 can be accepted.')
    
    return t_value


def t_test_diff(xbar_1, xbar_2, var_1, var_2, sample_size_1, sample_size_2, significance, one_tail=True):
    
    common_var = (((sample_size_1 - 1)*var_1) + ((sample_size_2 - 1)*var_2)) / (sample_size_1 + sample_size_2 - 2)
    
    t_value = (xbar_1 - xbar_2) / (np.sqrt(common_var)*np.sqrt((1/sample_size_1) + (1/sample_size_2)))

    dof = sample_size_1 + sample_size_2 - 2
        
    if one_tail & (t_value >= 0):
        
        p_value = 1 - stats.t.cdf(t_value, dof)
        
        t_alpha = stats.t.ppf(significance,dof)*(-1)
        
    elif one_tail & (t_value < 0):

        p_value = stats.t.cdf(t_value, dof) 
        
        t_alpha = stats.t.ppf(significance,dof)
        
    if one_tail == False:
        
        p_value = 2*(1 - stats.t.cdf(np.abs(t_value)))
        
        t_alpha = stats.t.ppf(significance/2,dof)*(-1)
    
    if p_value > significance:
        print(f't value = {t_value:.5f}')
        print(f't alpha = {t_alpha:.5f}')
        print(f'p value = {p_value:.5f} > {significance} and H0 can not be rejected.')
    else:
        print(f't value = {t_value:.5f}')
        print(f't alpha = {t_alpha:.5f}')
        print(f'p value = {p_value:.5f} <= {significance} and H0 is rejected, H1 can be accepted.')
    
    return t_value


def z_test(mu, var, xbar, sample_size, significance, one_tail=True):
    
    z_value = (xbar - mu) / np.sqrt(var/sample_size)
    
    if one_tail & (z_value >= 0):
        
        p_value = 1 - stats.norm.cdf(z_value)
        
        z_alpha = stats.norm.ppf(significance)*(-1)
        
    elif one_tail & (z_value < 0):

        p_value = stats.norm.cdf(z_value) 
        
        z_alpha = stats.norm.ppf(significance)
        
    if one_tail == False:
        
        p_value = 2*(1 - stats.norm.cdf(np.abs(z_value)))
        
        z_alpha = stats.norm.ppf(significance/2)*(-1)
    
    if p_value > significance:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} > {significance} and H0 can not be rejected.')
    else:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} <= {significance} and H0 is rejected, H1 can be accepted.')
    
    return z_value



def z_test_ratio(p, p_hat, sample_size, significance, one_tail=True):
    
    z_value = (p_hat - p) / np.sqrt(p*(1-p)/sample_size)
    
    if one_tail & (z_value >= 0):
        
        p_value = 1 - stats.norm.cdf(z_value)
        
        z_alpha = stats.norm.ppf(significance)*(-1)
        
    elif one_tail & (z_value < 0):

        p_value = stats.norm.cdf(z_value) 
        
        z_alpha = stats.norm.ppf(significance)
        
    if one_tail == False:
        
        p_value = 2*(1 - stats.norm.cdf(np.abs(z_value)))
        
        z_alpha = stats.norm.ppf(significance/2)*(-1)
    
    if p_value > significance:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} > {significance} and H0 can not be rejected.')
    else:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} <= {significance} and H0 is rejected, H1 can be accepted.')
    
    return z_value



def z_test_ratio_diff(p_hat, p_hat_1, p_hat_2, sample_size_1, sample_size_2, significance, one_tail=True):
    
    z_value = (p_hat_1 - p_hat_2) / np.sqrt(p_hat*(1-p_hat)*((1/sample_size_1)+(1/sample_size_2)))
    
    if one_tail & (z_value >= 0):
        
        p_value = 1 - stats.norm.cdf(z_value)
        
        z_alpha = stats.norm.ppf(significance)*(-1)
        
    elif one_tail & (z_value < 0):

        p_value = stats.norm.cdf(z_value) 
        
        z_alpha = stats.norm.ppf(significance)
        
    if one_tail == False:
        
        p_value = 2*(1 - stats.norm.cdf(np.abs(z_value)))
        
        z_alpha = stats.norm.ppf(significance/2)*(-1)
    
    if p_value > significance:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} > {significance} and H0 can not be rejected.')
    else:
        print(f'z value = {z_value:.5f}')
        print(f'z alpha = {z_alpha:.5f}')
        print(f'p value = {p_value:.5f} <= {significance} and H0 is rejected, H1 can be accepted.')
    
    return z_value
```
